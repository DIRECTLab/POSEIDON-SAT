# Analysis of ShipRSImageNet and Augmented Datasets

```python
import pandas as pd
from shiprsimagenet import ShipRSImageNet
```

## Ship Class Distributions

```python
dataset_stats = pd.DataFrame()
for dataset_name in 'ShipRSImageNet_V1', 'ShipRSImageNet_V1_Augmented', 'ShipRSImageNet_V1_Augmented_MatchedRes':
    dataset = ShipRSImageNet(dataset_name)
    objects = [ obj for img in dataset.get_image_set('train') for obj in img.objects ]
    counts = pd.Series([ obj.name for obj in objects ]).value_counts().rename(dataset.name)
    dataset_stats = pd.concat([dataset_stats, counts], axis=1)

dataset_stats.plot.bar(figsize=(12, 2));
```
